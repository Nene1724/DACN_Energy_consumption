## 2. Production-grade feature engineering from the old notebook

To preserve the current best production model, this notebook reuses the exact feature set from the old notebook and `EnergyPredictorService`.

Properties of the legacy feature set:

- it focuses on derived features directly related to inference cost,
- it includes `batch_high_power` to separate Jetson power profiles,
- it avoids categorical encoding to remain compatible with the runtime service.


```python
BASE_FEATURES_OLD = [
    "params_m",
    "gflops",
    "gmacs",
    "size_mb",
    "latency_avg_s",
    "throughput_iter_per_s",
]

FEATURE_COLUMNS_OLD = BASE_FEATURES_OLD + [
    "gflops_per_param",
    "gmacs_per_mb",
    "latency_throughput_ratio",
    "compute_intensity",
    "model_complexity",
    "computational_density",
    "log_params_m",
    "log_gflops",
    "log_size_mb",
    "log_gmacs",
    "log_latency",
    "log_throughput",
    "log_model_complexity",
    "log_compute_intensity",
    "log_params_x_log_latency",
    "log_gflops_x_log_latency",
    "batch_high_power",
]

JETSON_BATCH_THRESHOLD = 10.0

def smape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = np.abs(y_true) + np.abs(y_pred) + 1e-9
    return float(np.mean(2.0 * np.abs(y_pred - y_true) / denom) * 100.0)

def mape_eps(y_true, y_pred, eps=1e-6):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = np.maximum(np.abs(y_true), eps)
    return float(np.mean(np.abs((y_true - y_pred) / denom)) * 100.0)

def regression_metrics(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return {
        "mae": float(np.mean(np.abs(y_true - y_pred))),
        "rmse": float(np.sqrt(np.mean((y_true - y_pred) ** 2))),
        "smape": float(smape(y_true, y_pred)),
        "mape_eps": float(mape_eps(y_true, y_pred)),
        "r2_original": float(__import__("sklearn.metrics").metrics.r2_score(y_true, y_pred)),
        "r2_log": float(__import__("sklearn.metrics").metrics.r2_score(np.log1p(y_true), np.log1p(np.clip(y_pred, 0, None)))),
        "median_abs_error": float(np.median(np.abs(y_true - y_pred))),
    }

def add_old_device_features(df: pd.DataFrame, device_key: str) -> pd.DataFrame:
    df = df.copy()
    df["device"] = device_key
    df["gflops_per_param"] = df["gflops"] / (df["params_m"] + 1e-6)
    df["gmacs_per_mb"] = df["gmacs"] / (df["size_mb"] + 1e-6)
    df["latency_throughput_ratio"] = df["latency_avg_s"] * df["throughput_iter_per_s"]
    df["compute_intensity"] = df["gflops"] * df["latency_avg_s"]
    df["model_complexity"] = df["params_m"] * df["gflops"]
    df["computational_density"] = df["gflops"] / (df["size_mb"] + 1e-6)
    df["log_params_m"] = np.log1p(df["params_m"])
    df["log_gflops"] = np.log1p(df["gflops"])
    df["log_size_mb"] = np.log1p(df["size_mb"])
    df["log_gmacs"] = np.log1p(df["gmacs"])
    df["log_latency"] = np.log1p(df["latency_avg_s"])
    df["log_throughput"] = np.log1p(df["throughput_iter_per_s"])
    df["log_model_complexity"] = np.log1p(df["model_complexity"])
    df["log_compute_intensity"] = np.log1p(df["compute_intensity"])
    df["log_params_x_log_latency"] = df["log_params_m"] * df["log_latency"]
    df["log_gflops_x_log_latency"] = df["log_gflops"] * df["log_latency"]
    if "jetson" in device_key:
        power_ratio = df["energy_avg_mwh"] / (df["latency_avg_s"] + 1e-9)
        df["batch_high_power"] = (power_ratio > JETSON_BATCH_THRESHOLD).astype(int)
    else:
        df["batch_high_power"] = 0
    return df

legacy_df = pd.concat(
    [
        add_old_device_features(research_df[research_df["device_type"] == "jetson_nano"], "jetson_nano"),
        add_old_device_features(research_df[research_df["device_type"] == "raspberry_pi5"], "raspberry_pi5"),
    ],
    ignore_index=False,
).sort_index()

print("Legacy/production dataframe shape:", legacy_df.shape)
display(legacy_df[FEATURE_COLUMNS_OLD + ["device_type", "energy_avg_mwh"]].head())
```

## 5. EDA: model-family and function-group distributions

This section is highly useful for research reporting because it shows:

- which families dominate the dataset,
- how balanced the architecture coverage is,
- how `mobile_edge`, `cnn_backbone`, `vision_transformer`, and related groups differ.


```python
def wilson_interval(k: int, n: int, z: float = 1.96):
    if n <= 0:
        return 0.0, 0.0
    p = k / n
    denom = 1.0 + (z ** 2) / n
    center = (p + (z ** 2) / (2 * n)) / denom
    margin = z * np.sqrt((p * (1 - p) + (z ** 2) / (4 * n)) / n) / denom
    return max(0.0, center - margin), min(1.0, center + margin)

def format_family_label(label: str) -> str:
    tokens = label.replace("_", " ").split()
    if not tokens:
        return label
    replacements = {
        "vit": "ViT",
        "deit": "DeiT",
        "convnext": "ConvNeXt",
        "regnet": "RegNet",
        "mnasnet": "MnasNet",
        "efficientnet": "EfficientNet",
    }
    head = tokens[0].lower()
    if head in replacements:
        tokens[0] = replacements[head]
        return " ".join(tokens)
    return " ".join(word.capitalize() for word in tokens)

def format_function_label(label: str) -> str:
    mapping = {
        "vision_transformer": "ViT",
        "cnn_backbone": "CNN backbone",
        "mobile_edge": "Mobile edge",
        "object_detection": "Detection",
        "pose_estimation": "Pose",
        "segmentation": "Segmentation",
        "classification": "Classification",
        "super_resolution": "Super-res",
    }
    if label in mapping:
        return mapping[label]
    return label.replace("_", " ").title()

def prepare_topk_percent(df: pd.DataFrame, category_col: str, device_col: str, top_k: int = 8):
    counts = df.groupby([device_col, category_col]).size().reset_index(name="count")
    device_total = df.groupby(device_col).size().rename("device_total")
    counts = counts.merge(device_total, on=device_col, how="left")
    total_counts = counts.groupby(category_col)["count"].sum().sort_values(ascending=False)
    top_categories = total_counts.head(top_k).index
    counts["category"] = counts[category_col].where(counts[category_col].isin(top_categories), "Other")
    agg = counts.groupby([device_col, "category"], as_index=False).agg(
        count=("count", "sum"),
        device_total=("device_total", "first"),
    )
    agg["percent"] = agg["count"] / agg["device_total"]
    ci_bounds = agg.apply(lambda row: wilson_interval(int(row["count"]), int(row["device_total"])), axis=1)
    agg["ci_low"] = [b[0] for b in ci_bounds]
    agg["ci_high"] = [b[1] for b in ci_bounds]
    agg["percent_pct"] = agg["percent"] * 100.0
    agg["ci_low_pct"] = agg["ci_low"] * 100.0
    agg["ci_high_pct"] = agg["ci_high"] * 100.0
    return agg

def order_by_device(agg: pd.DataFrame, device_label: str, other_label: str = "Other") -> list:
    pivot = agg.pivot_table(index="category", columns="device_label", values="percent_pct", aggfunc="max").fillna(0)
    if device_label not in pivot.columns:
        device_label = pivot.columns[0]
    ordered = pivot.sort_values(device_label, ascending=False).index.tolist()
    if other_label in ordered:
        ordered.remove(other_label)
        ordered.append(other_label)
    return ordered

device_label_map = {
    "jetson_nano": "Jetson Nano",
    "raspberry_pi5": "Raspberry Pi 5",
    "Jetson Nano": "Jetson Nano",
    "Raspberry Pi 5": "Raspberry Pi 5",
}

family_df = prepare_topk_percent(
    research_df, category_col="model_family", device_col="device_name", top_k=10
 )
family_df["device_label"] = family_df["device_name"].map(device_label_map).fillna(family_df["device_name"])
family_df["category_label"] = family_df["category"].apply(format_family_label)

function_df = prepare_topk_percent(
    research_df, category_col="function_group", device_col="device_name", top_k=8
 )
function_df["device_label"] = function_df["device_name"].map(device_label_map).fillna(function_df["device_name"])
function_df["category_label"] = function_df["category"].apply(format_function_label)

baseline_device = "Raspberry Pi 5"
family_order = order_by_device(family_df, baseline_device)
function_order = order_by_device(function_df, baseline_device)

palette = sns.color_palette("colorblind", n_colors=2)
hue_order = sorted(family_df["device_label"].unique())

fig, axes = plt.subplots(1, 2, figsize=(22, 9))
bar_height = 0.32
offsets = np.linspace(-bar_height / 2, bar_height / 2, len(hue_order))
x_max = 55.0

def pick_highlight(order: list) -> str:
    for name in order:
        if name != "Other":
            return name
    return order[0] if order else ""

def plot_horizontal_percent(ax, data, order, label_column):
    y_pos = np.arange(len(order))
    highlight_label = pick_highlight(order)
    for i, device in enumerate(hue_order):
        subset = data[data["device_label"] == device].set_index("category").reindex(order).fillna(0)
        values = subset["percent_pct"].to_numpy()
        ci_low = subset["ci_low_pct"].to_numpy()
        ci_high = subset["ci_high_pct"].to_numpy()
        err = np.vstack([values - ci_low, ci_high - values])
        bars = ax.barh(
            y_pos + offsets[i],
            values,
            height=bar_height / max(1, len(hue_order)),
            color=palette[i],
            label=device,
            edgecolor="0.2",
            linewidth=0.6,
        )
        ax.errorbar(
            values,
            y_pos + offsets[i],
            xerr=err,
            fmt="none",
            ecolor="0.25",
            elinewidth=1.0,
            capsize=3,
            capthick=1.0,
            zorder=3,
        )
        for bar, value in zip(bars, values):
            if value <= 0:
                continue
            if value >= x_max - 6:
                x_text = value - 0.8
                ha = "right"
                color = "white"
            else:
                x_text = value + 0.6
                ha = "left"
                color = "0.2"
            ax.text(
                x_text,
                bar.get_y() + bar.get_height() / 2,
                f"{value:.1f}%",
                va="center",
                ha=ha,
                fontsize=9,
                color=color,
            )
    labels = data.drop_duplicates("category").set_index("category")[label_column].reindex(order).tolist()
    labels = ["★ " + label if name == highlight_label else label for name, label in zip(order, labels)]
    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels, fontsize=11)
    ax.set_xlabel("Share of models within each device (%)", fontsize=12)
    ax.set_xlim(0, x_max)
    ax.grid(axis="x", alpha=0.2)

plot_horizontal_percent(
    axes[0],
    family_df,
    family_order,
    "category_label",
)
plot_horizontal_percent(
    axes[1],
    function_df,
    function_order,
    "category_label",
)

legend = axes[0].legend(
    title="Device",
    loc="upper center",
    bbox_to_anchor=(1.02, 1.12),
    ncol=2,
    frameon=False,
)
legend.get_title().set_fontsize(11)

despine_axes(axes)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig(FIGURE_DIR / "02_family_function_distribution.png", dpi=300, bbox_inches="tight")
plt.show()

display(
    research_df.groupby(["device_name", "function_group"]).size().reset_index(name="samples").sort_values(
        ["device_name", "samples"], ascending=[True, False]
    )
)
```

## 9. Retrain the best production model from the old notebook

Here we preserve the original production philosophy:

- **Jetson**: tune for `RMSE` using stratified cross-validation over energy bins
- **Raspberry Pi 5**: tune for `SMAPE`
- the core model remains **Extra Trees + log1p target**


```python
def build_ttr_pipeline(regressor):
    return Pipeline(
        [
            ("scaler", RobustScaler()),
            (
                "model",
                TransformedTargetRegressor(
                    regressor=regressor,
                    func=np.log1p,
                    inverse_func=np.expm1,
                    check_inverse=False,
                ),
            ),
        ]
    )

SMAPE_SCORER = make_scorer(smape, greater_is_better=False)

def tune_jetson_model(df: pd.DataFrame):
    X = df[FEATURE_COLUMNS_OLD]
    y = df["energy_avg_mwh"].values
    bins = KBinsDiscretizer(
        n_bins=10,
        encode="ordinal",
        strategy="quantile",
        subsample=None,
        quantile_method="averaged_inverted_cdf",
    ).fit_transform(y.reshape(-1, 1)).ravel().astype(int)
    cv = list(StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED).split(X, bins))

    pipeline = build_ttr_pipeline(ExtraTreesRegressor(random_state=SEED, n_jobs=-1))
    param_dist = {
        "model__regressor__n_estimators": randint(800, 1401),
        "model__regressor__max_depth": [12, 16, 20, 24, 28, None],
        "model__regressor__min_samples_split": randint(2, 6),
        "model__regressor__min_samples_leaf": randint(1, 3),
        "model__regressor__max_features": [0.7, 0.8, 0.9, 1.0],
        "model__regressor__bootstrap": [False],
    }

    search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=param_dist,
        n_iter=35,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        random_state=SEED,
        n_jobs=-1,
        verbose=0,
    )
    search.fit(X, y)
    pred = cross_val_predict(search.best_estimator_, X, y, cv=cv, n_jobs=1)
    metrics = regression_metrics(y, pred)
    return {
        "device_key": "jetson_nano",
        "best_estimator": search.best_estimator_.fit(X, y),
        "best_params": search.best_params_,
        "cv_metrics": metrics,
        "cv_predictions": pred,
        "samples": int(len(df)),
        "cv_strategy": "StratifiedKFold(n_splits=10)",
    }

def tune_rpi5_model(df: pd.DataFrame):
    X = df[FEATURE_COLUMNS_OLD]
    y = df["energy_avg_mwh"].values
    cv = KFold(n_splits=5, shuffle=True, random_state=SEED)

    pipeline = build_ttr_pipeline(ExtraTreesRegressor(random_state=SEED, n_jobs=-1))
    param_dist = {
        "model__regressor__n_estimators": randint(900, 1501),
        "model__regressor__max_depth": [8, 10, 12, 14, 16, 18],
        "model__regressor__min_samples_split": randint(2, 5),
        "model__regressor__min_samples_leaf": randint(1, 3),
        "model__regressor__max_features": [0.6, 0.7, 0.8, 0.9],
        "model__regressor__bootstrap": [False],
    }

    search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=param_dist,
        n_iter=25,
        scoring=SMAPE_SCORER,
        cv=cv,
        random_state=SEED,
        n_jobs=-1,
        verbose=0,
    )
    search.fit(X, y)
    pred = cross_val_predict(search.best_estimator_, X, y, cv=cv, n_jobs=1)
    metrics = regression_metrics(y, pred)
    return {
        "device_key": "raspberry_pi5",
        "best_estimator": search.best_estimator_.fit(X, y),
        "best_params": search.best_params_,
        "cv_metrics": metrics,
        "cv_predictions": pred,
        "samples": int(len(df)),
        "cv_strategy": "KFold(n_splits=5)",
    }

legacy_jetson_df = legacy_df[legacy_df["device_type"] == "jetson_nano"].copy()
legacy_rpi5_df = legacy_df[legacy_df["device_type"] == "raspberry_pi5"].copy()

jetson_result = tune_jetson_model(legacy_jetson_df)
rpi5_result = tune_rpi5_model(legacy_rpi5_df)

display(pd.DataFrame([jetson_result["cv_metrics"]]).assign(device="Jetson Nano"))
display(pd.DataFrame([rpi5_result["cv_metrics"]]).assign(device="Raspberry Pi 5"))
```

## 11. Detailed analysis of the production winner

If the old production model still wins, we analyze it in detail for each device:

- actual vs predicted,
- residual distribution,
- the `batch_high_power` behavior on Jetson,
- and the hardest samples to predict.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def summarize_metrics(actual: np.ndarray, predicted: np.ndarray):
    return {
        "r2": r2_score(actual, predicted),
        "mae": mean_absolute_error(actual, predicted),
        "rmse": np.sqrt(mean_squared_error(actual, predicted)),
    }

def percentile_clip(values: np.ndarray, low: float = 1.0, high: float = 99.0):
    return np.percentile(values, [low, high])

def outlier_count(values: np.ndarray, low: float, high: float) -> int:
    return int(((values < low) | (values > high)).sum())

jetson_diag = pd.DataFrame({
    "actual": legacy_jetson_df["energy_avg_mwh"].values,
    "predicted": jetson_result["cv_predictions"],
})
jetson_diag["residual"] = jetson_diag["predicted"] - jetson_diag["actual"]
jetson_diag["abs_pct_error"] = np.abs(jetson_diag["residual"]) / np.maximum(np.abs(jetson_diag["actual"]), 1e-9) * 100.0

rpi5_diag = pd.DataFrame({
    "actual": legacy_rpi5_df["energy_avg_mwh"].values,
    "predicted": rpi5_result["cv_predictions"],
})
rpi5_diag["residual"] = rpi5_diag["predicted"] - rpi5_diag["actual"]
rpi5_diag["abs_pct_error"] = np.abs(rpi5_diag["residual"]) / np.maximum(np.abs(rpi5_diag["actual"]), 1e-9) * 100.0

jetson_metrics = summarize_metrics(jetson_diag["actual"], jetson_diag["predicted"])
rpi5_metrics = summarize_metrics(rpi5_diag["actual"], rpi5_diag["predicted"])

jetson_min, jetson_max = percentile_clip(
    np.concatenate([jetson_diag["actual"], jetson_diag["predicted"]]), low=0.5, high=99.5
 )
rpi5_min, rpi5_max = percentile_clip(
    np.concatenate([rpi5_diag["actual"], rpi5_diag["predicted"]]), low=0.5, high=99.5
 )

jetson_res_low, jetson_res_high = percentile_clip(jetson_diag["residual"], low=1, high=99)
rpi5_res_low, rpi5_res_high = percentile_clip(rpi5_diag["residual"], low=1, high=99)

jetson_outliers = outlier_count(jetson_diag["residual"], jetson_res_low, jetson_res_high)
rpi5_outliers = outlier_count(rpi5_diag["residual"], rpi5_res_low, rpi5_res_high)

fig, axes = plt.subplots(2, 2, figsize=(18, 14))

scatter_style = dict(alpha=0.3, s=40, edgecolor="none")
metric_box = dict(boxstyle="round,pad=0.35", fc="white", ec="0.6", alpha=0.9)

jetson_min_plot = max(jetson_min, 1e-2)
jetson_max_plot = jetson_max
rpi5_min_plot = max(rpi5_min, 0.0)
rpi5_max_plot = rpi5_max

# Jetson Nano: Actual vs Predicted (log scale + aligned axes)
axes[0, 0].scatter(jetson_diag["actual"], jetson_diag["predicted"], color="#1f77b4", **scatter_style)
axes[0, 0].plot([jetson_min_plot, jetson_max_plot], [jetson_min_plot, jetson_max_plot], linestyle="--", color="black", lw=1)
axes[0, 0].set_xscale("log")
axes[0, 0].set_yscale("log")
axes[0, 0].set_xlim(jetson_min_plot, jetson_max_plot)
axes[0, 0].set_ylim(jetson_min_plot, jetson_max_plot)
axes[0, 0].set_xlabel("Actual energy (mWh)")
axes[0, 0].set_ylabel("Predicted energy (mWh)")
axes[0, 0].text(
    0.03,
    0.97,
    f"R2={jetson_metrics['r2']:.3f}\nMAE={jetson_metrics['mae']:.1f} mWh\nRMSE={jetson_metrics['rmse']:.1f} mWh",
    transform=axes[0, 0].transAxes,
    ha="left",
    va="top",
    fontsize=10,
    bbox=metric_box,
 )

# Jetson Nano: Residual distribution (clipped view)
jetson_residuals_clipped = jetson_diag["residual"].clip(jetson_res_low, jetson_res_high)
sns.histplot(jetson_residuals_clipped, kde=True, color="#1f77b4", ax=axes[0, 1])
axes[0, 1].axvline(0, linestyle="--", color="black", lw=1)
axes[0, 1].set_xlabel("Residual (mWh)")
axes[0, 1].set_ylabel("Density")
axes[0, 1].set_xlim(jetson_res_low, jetson_res_high)
axes[0, 1].text(
    0.03,
    0.95,
    f"Outliers outside view: {jetson_outliers}",
    transform=axes[0, 1].transAxes,
    ha="left",
    va="top",
    fontsize=10,
    bbox=metric_box,
 )

# Raspberry Pi 5: Actual vs Predicted (aligned axes)
axes[1, 0].scatter(rpi5_diag["actual"], rpi5_diag["predicted"], color="#d62728", **scatter_style)
axes[1, 0].plot([rpi5_min_plot, rpi5_max_plot], [rpi5_min_plot, rpi5_max_plot], linestyle="--", color="black", lw=1)
axes[1, 0].set_xlim(rpi5_min_plot, rpi5_max_plot)
axes[1, 0].set_ylim(rpi5_min_plot, rpi5_max_plot)
axes[1, 0].set_xlabel("Actual energy (mWh)")
axes[1, 0].set_ylabel("Predicted energy (mWh)")
axes[1, 0].text(
    0.03,
    0.97,
    f"R2={rpi5_metrics['r2']:.3f}\nMAE={rpi5_metrics['mae']:.2f} mWh\nRMSE={rpi5_metrics['rmse']:.2f} mWh",
    transform=axes[1, 0].transAxes,
    ha="left",
    va="top",
    fontsize=10,
    bbox=metric_box,
 )

# Raspberry Pi 5: Residual distribution (clipped view)
rpi5_residuals_clipped = rpi5_diag["residual"].clip(rpi5_res_low, rpi5_res_high)
sns.histplot(rpi5_residuals_clipped, kde=True, color="#d62728", ax=axes[1, 1])
axes[1, 1].axvline(0, linestyle="--", color="black", lw=1)
axes[1, 1].set_xlabel("Residual (mWh)")
axes[1, 1].set_ylabel("Density")
axes[1, 1].set_xlim(rpi5_res_low, rpi5_res_high)
axes[1, 1].text(
    0.03,
    0.95,
    f"Outliers outside view: {rpi5_outliers}",
    transform=axes[1, 1].transAxes,
    ha="left",
    va="top",
    fontsize=10,
    bbox=metric_box,
 )

for ax in axes.flat:
    ax.grid(True, alpha=0.2)

despine_axes(axes)
plt.tight_layout()
plt.savefig(FIGURE_DIR / "07_production_model_diagnostics.png", dpi=300, bbox_inches="tight")
plt.show()

display(jetson_diag.describe().round(4))
display(rpi5_diag.describe().round(4))
```
